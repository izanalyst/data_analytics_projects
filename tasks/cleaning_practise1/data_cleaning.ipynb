## Data Cleaning project

## 📊 Dataset
List of highest grossing concert tours by Women

## Goal:
Clean raw data and prepare it for analysis

## 🛠 Tools
- Python
- Pandas

## 🧹 Cleaning Steps:

## 1. import pandas and read csv file

```python
import pandas as pd
import re

df = pd.read_csv("data\my_file.csv")
```

## 2. dropping of unwanted columns, removing extra symbols, reference markers ([1], [2] and etc) and whitespaces

```python
df = df.drop(columns=["Ref."])
df = df.replace(r'\[[a-zA-Z0-9]+\]', '', regex=True)
df = df.replace(r'[^\w \–\&]', '', regex=True)
df.columns = df.columns.str.replace('\xa0', ' ').str.strip()
```

## 3. Cleaning monetary values ($, commas) and converting columns to numeric types

```python
money_columns = ["Actual gross", "Adjusted gross (in 2022 dollars)", "Average gross"]

for col in money_columns:
    if col in df.columns:
        df[col] = (
            df[col].astype(str).str.replace(r'[\$,]', '', regex=True)
        )
        df[col] = pd.to_numeric(df[col], errors="coerce")

numeric_columns = ['Peak', 'All Time Peak', 'Shows']

for col in numeric_columns:
    if col in df.columns:
        df[col] = df[col].astype("Int64")
```

## 4. Processed year column into start and end year

```python
df[['start_year', 'end_year']] = df['Year(s)'].str.split('–', expand=True)
df['start_year'] = pd.to_numeric(df['start_year'], errors='coerce')
df['end_year'] = pd.to_numeric(df['end_year'], errors='coerce')

df['end_year'] = df['end_year'].fillna(df['start_year']).astype("Int64")
```

## 5. 📈 Result
Data is cleaned and ready for analysis

```python
print(df.isnull().sum())
df.info()

df.to_csv("data\cleaned_data.csv", index=False)
print(df)
```
